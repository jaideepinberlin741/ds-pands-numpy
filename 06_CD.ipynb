Pandas functions that allow us to combine two sets of data include the use of `pd.merge()`, `df.join()`, `df.merge()`, and `pd.concat()`. For the most part, these do largely the same things (although you'll notice the slight syntax difference with `merge()` and `concat()` being able to be called via the pandas module and `merge()` and `join()` being able to be called on a DataFrame instance). There are some cases where one of these might be better than another in terms of writing less code or performing some kind of data combination in an easier way. The major differences between these, though, largely depend on what they do by default when you try to combine different data. By default, `merge()` looks to join on common columns, `join()` on common indices, and `concat()` by just appending on a given axis.

You can find more detail about the differences between all three of these in the [docs](https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html). We'll look at some examples below. 

`````python
# We'll go back to our wine data set. Who doesn't love wine?
import pandas as pd
wine_df = pd.read_csv('data/winequality-red.csv', delimiter=';')
wine_df.head()
`````

`````python
# A glance at the values of the quality of wine in the DataFrame
wine_df.quality.unique()
`````

`````python
# get_dummies is a method called on the pandas module - you simply pass in a pandas Series 
# or DataFrame, and it will convert a categorical variable into dummy/indicator variables. 
quality_dummies = pd.get_dummies(wine_df.quality, prefix='quality')
quality_dummies.head()
`````

Now let's look at the `join()` method. Remember, this joins on indices by default. This means that we can simply join our quality dummies DataFrame back to our original wine DataFrame with the following...

`````python
joined_df = wine_df.join(quality_dummies)
joined_df.head() 
`````

The arguments of `.join` are the following: 
````
DataFrame.join(self, other, on=None, how='left', lsuffix='', rsuffix='', sort=False)
````
With `how` we can specify which join method we want to use.

The how argument to merge specifies how to determine which keys are to be included in the resulting table. If a key combination does not appear in either the left or right tables, the values in the joined table will be NA. Here is a summary of the how options and their SQL equivalent names:

Merge method | SQL Join Name | Description
---|---|---
left| LEFT OUTER JOIN | Use keys from left frame only
right | RIGHT OUTER JOIN | Use keys from right frame only
outer | FULL OUTER JOIN | Use union of keys from both frames
inner | INNER JOIN | Use intersection of keys from both frames


You can also think of it as set theory and use Venn diagrams to illustrate what happens in each method.

2. You have received some weather data (temperature) of the last year. For each month the average temperature was measured, only for a few months the maximum temperature could be measured. Anyway, you want to combine these two data without losing any information.

(Extra question: Can you fill in the average max. Temperature for the missing values in the Column `Max TemperatureF`)

`````python
weather_mean_data = {'Mean TemperatureF': [53.1, 70., 34.93548387, 28.71428571, 32.35483871, 72.87096774, 70.13333333, 35., 62.61290323, 39.8, 55.4516129 , 63.76666667],
                     'Month': ['Apr', 'Aug', 'Dec', 'Feb', 'Jan', 'Jul', 'Jun', 'Mar', 'May', 'Nov', 'Oct', 'Sep']}
weather_max_data = {'Max TemperatureF': [68, 89, 91, 84], 'Month': ['Jan', 'Apr', 'Jul', 'Oct']}
`````

`````python
df1 = pd.DataFrame(weather_mean_data)
df2 = pd.DataFrame(weather_max_data)
df1.join(df2.set_index('Month'), on='Month', how='outer', lsuffix='_left', rsuffix='_right')
`````

`````python
# pandas provides DataFrame, a 2D table-like structure similar to a spreadsheet.
import pandas as pd

# Imports the LinearRegression class from scikit-learn.
# This class implements ordinary least squares linear regression (fits y = a * x + b).
from sklearn.linear_model import LinearRegression

# Imports NumPy for numeric arrays.
# In this script np is not strictly used, but imported often by habit.
import numpy as np

# --- INPUT DATA ---
# Data definition (dictionaries)
weather_mean_data = {
    'Mean TemperatureF': [53.1, 70., 34.93548387, 28.71428571, 32.35483871,
                          72.87096774, 70.13333333, 35., 62.61290323, 39.8,
                          55.4516129, 63.76666667],
    'Month': ['Apr', 'Aug', 'Dec', 'Feb', 'Jan', 'Jul', 'Jun', 'Mar',
              'May', 'Nov', 'Oct', 'Sep']
}

weather_max_data = {
    'Max TemperatureF': [68, 89, 91, 84],
    'Month': ['Jan', 'Apr', 'Jul', 'Oct']
}

# --- CREATE DATAFRAMES ---
# Convert dictionaries to DataFrames
df_mean = pd.DataFrame(weather_mean_data)
df_max = pd.DataFrame(weather_max_data)

# --- MERGE (lossless) ---
# Merge the two tables on 'Month' using a left join
# merge combines rows from df_mean and df_max by matching the values in the "Month" column.
# on="Month": match rows where the Month values are equal.
# how="left": left-join — keep all rows from the left table (df_mean). If there is a matching row in df_max, its Max TemperatureF is added. If not, Max TemperatureF will be NaN (pandas’ "missing value").
# Why left? Because you want to keep all months for which you have mean temperatures and just attach max where available.
df = df_mean.merge(df_max, on="Month", how="left")

# Prints the merged table to the console so you can inspect it.
# Good practice when learning.
print("Merged Data:")
print(df)

# ----------------------------------------------------------------
# --- FILL MISSING VALUES IN 'Max TemperatureF' USING REGRESSION ---
# ----------------------------------------------------------------

# Select rows where Max TemperatureF IS known
# dropna(subset=[...]) returns a new DataFrame with rows that have non-missing values in the specified column(s).
# known now contains only the rows where we measured Max TemperatureF.
# These rows will be used to train the regression model.
# Why? You can only train a supervised model on rows where both input (Mean) and target (Max) are present.
known = df.dropna(subset=["Max TemperatureF"])

# Prepare X (features) and y (target) for training
# X_train is a DataFrame with one column: Mean TemperatureF.
# y_train is a Series with the corresponding Max TemperatureF values.
# X_train is the input matrix (features) used to predict.
# Here we use only one feature: the mean temperature.
# Note the double bracket [[...]] — this creates a DataFrame with a single column (shape (n_samples, 1)), which scikit-learn expects.
# y_train is the target vector (the values to predict):
# known max temperatures. y_train is a pandas Series (shape (n_samples,)).
# Important: scikit-learn expects X to be 2-dimensional (even when you have one feature) and y can be 1-dimensional.
X_train = known[["Mean TemperatureF"]]
y_train = known["Max TemperatureF"]

# Train linear model
# Create the regression model and train it on the known data.
# model = LinearRegression() creates an instance of the linear regression model.
# model.fit(X_train, y_train) fits the line Max ≈ a * Mean + b to the training points by minimizing squared errors (ordinary least squares).
# After .fit(), the model has attributes:
# model.coef_ — slope(s) a (array).
# model.intercept_ — intercept b (scalar).
# model.score(X_train, y_train) returns the R² on the training set (good to check fit quality).
model = LinearRegression()
model.fit(X_train, y_train)

# You could print these if you want to inspect the fit:
import ipywidgets as widgets
from IPython.display import display

# Creation of buttons
button_yes = widgets.Button(description="Yes")
button_no = widgets.Button(description="No")

output = widgets.Output()

def on_button_yes(b):
    with output:
        print("Coef:", model.coef_)
        print("Intercept:", model.intercept_)
        print("R^2 on training:", model.score(X_train, y_train))

def on_button_no(b):
    with output:
        print("Model details not shown.")

button_yes.on_click(on_button_yes)
button_no.on_click(on_button_no)

display(widgets.Label("Show model details?"))
display(button_yes, button_no, output)


# Predict missing values
# Select rows where Max is missing (to predict)
# df["Max TemperatureF"].isna() returns a boolean Series: True for rows where Max is missing.
# df[ ... ] filters the DataFrame to only those rows.
# [[ "Mean TemperatureF" ]] selects only the feature column;
# result is a DataFrame shaped (n_missing, 1) ready for model.predict.
missing = df[df["Max TemperatureF"].isna()][["Mean TemperatureF"]]
# Predict missing Max values
# model.predict(...) returns a NumPy array of predicted Max TemperatureF values corresponding to each row in missing.
# The length of predicted_values equals number of missing rows.
predicted_values = model.predict(missing)

# Fill predicted values into DataFrame
# Insert predicted values into original DataFrame
# .loc[row_selector, column_name] = values assigns values to selected cells.
# df["Max TemperatureF"].isna() selects the rows where Max was missing. We assign the predicted_values into the "Max TemperatureF" column for exactly those rows.
# pandas aligns assignment by position here because the boolean mask selects the exact rows in order and predicted_values is an array of matching length.
df.loc[df["Max TemperatureF"].isna(), "Max TemperatureF"] = predicted_values

# Print filled DataFrame
# Shows the table with estimated max values filled in.
print("\nFilled Data with Estimated Maximum Temperatures:")
print(df)

# OPTIONAL: Round values for nicer output
# Round values for nicer output
df["Max TemperatureF"] = df["Max TemperatureF"].round(1)
print("\nFinal Output:")
print(df)
`````
